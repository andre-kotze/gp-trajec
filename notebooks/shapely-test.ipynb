```python
import matplotlib.pyplot as plt
from shapely.geometry import LineString, Polygon, MultiPolygon, Point
import numpy as np
from numpy import cos, sin
```

```python
building = Polygon([(0,10),
                    (5,10),
                    (5,5),
                    (0,5)
                    ])

military = Polygon([(15,10),
                    (25,10),
                    (25,0)
                    ])

nofly = Polygon([(0,25),
                (15,25),
                (15,20),
                (0,20)
                ])

nofly = Polygon([(0,25,15),
                (0,20,15),
                (10,15,15),
                (15,20,15),
                (15,25,15)
                ])

barriers = [building, military, nofly]

for barrier in barriers:
    plt.plot(*barrier.exterior.xy)

intersection = nofly.intersects(military)

print(intersection)
```

```python
pts = [(0,25,15),
                (0,20,15),
                (10,15,15),
                (15,20,15),
                (15,25,15)
                ]
pts.extend([(tpl[0],tpl[1],0) for tpl in pts])
# flip coords/view
pts = [(tpl[0],tpl[2],tpl[1]) for tpl in pts]
nofly_s = Polygon(pts)

plt.plot(*nofly_s.exterior.xy)
```
